```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

import torch

from src.config import SMDConfig
from src.data.preprocessing import compute_smd_normalization_stats
```

Load dataset structure

```python
cfg = SMDConfig()

print("Dataset Root:", cfg.root_dir)
print("Train Files:", len(os.listdir(os.path.join(cfg.root_dir, "train"))))
print("Test Files:", len(os.listdir(os.path.join(cfg.root_dir, "test"))))
print("Label Files:", len(os.listdir(os.path.join(cfg.root_dir, "test_label"))))
```

Compute normalization stats

```python
mean, std = compute_smd_normalization_stats(cfg.root_dir)
print("Mean shape:", mean.shape)
print("Std shape:", std.shape)
```

Visualize a sample time-series

```python
sample_path = os.path.join(cfg.root_dir, "train", sorted(os.listdir(os.path.join(cfg.root_dir, "train")))[0])
df = pd.read_csv(sample_path, header=None, delim_whitespace=True).values

plt.figure(figsize=(12, 4))
plt.plot(df[:, 0], label="Feature 0")
plt.plot(df[:, 1], label="Feature 1")
plt.title("Raw Time Series Sample")
plt.legend()
plt.show()
```
